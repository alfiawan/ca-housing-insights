```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing

housing = fetch_california_housing()
df = pd.DataFrame(data=housing.data, columns=housing.feature_names)
df['target'] = housing.target

print("First 1o row of datasets")
print(df.head(10))

print("Last 10 row of datasets")
print(df.tail(10))

print("\nDescriptive statistics:")
print(df.describe())

print("\nData information:")
print(df.info())

plt.figure(figsize=(8, 6))  # Mengatur ukuran gambar
plt.hist(df['MedInc'], bins=30) # Membuat histogram # Changed 'median_income' to 'MedInc'
plt.xlabel('Median Income') # Memberi label sumbu x
plt.ylabel('Frequency') # Memberi label sumbu y
plt.title('Distribution of Median Income') # Memberi judul grafik
plt.show() # Menampilkan grafik

# Scatter plot 'median_house_value' vs 'MedInc'
plt.figure(figsize=(8, 6))
plt.scatter(df['MedInc'], df['target'], alpha=0.5) # alpha untuk transparansi # Changed 'median_house_value' to 'target'
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.title('Median House Value vs Median Income')
plt.show()

# Ulangi untuk fitur lain:
# Scatter plot 'median_house_value' vs 'AveRooms'
plt.figure(figsize=(8, 6))
plt.scatter(df['AveRooms'], df['target'], alpha=0.5) # Changed 'median_house_value' to 'target'
plt.xlabel('Average Rooms')
plt.ylabel('Median House Value')
plt.title('Median House Value vs Average Rooms')
plt.show()

df['house_value_per_room'] = df['target'] / df['AveRooms']

plt.figure(figsize=(8, 6))
plt.hist(df['house_value_per_room'], bins=30)
plt.xlabel('House Value per Room')
plt.ylabel('Frequency')
plt.title('Distribution of House Value per Room')
plt.show()

plt.figure(figsize=(8, 6))
plt.scatter(df['MedInc'], df['house_value_per_room'], alpha=0.5)
plt.xlabel('Median Income')
plt.ylabel('House Value per Room')
plt.title('House Value per Room vs Median Income')
plt.show()
```
